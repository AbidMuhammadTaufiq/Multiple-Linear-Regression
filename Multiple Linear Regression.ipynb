# Definisi
Analisis regresi digunakan untuk mengukur seberapa besar pengaruh antara variabel bebas dan variabel terikat. Regresi linear berganda merupakan model regresi yang melibatkan lebih dari satu variabel independen. Analisis regresi linear berganda dilakukan untuk mengetahui arah dan seberapa besar pengaruh variabel independen terhadap variabel dependen.


# Bentuk Umum

Bentuk umum persamaan regresi linear ganda adalah sebagai berikut
\begin{equation}
Y=b_0 + b_1 X_1 + b_2 X_2
\end{equation}

dengan 
ğ‘Œ = variabel independen
ğ‘‹ = variabel dependen
ğ‘ = koefisien regresi

    >>> dataX = {'x1' : [12, 14, 10, 16, 18, 24, 12], 'x2' : [10, 11, 14, 13, 15, 20, 8]}
    >>> dataY = [6, 7, 8, 8, 9, 10, 5]
    >>> b = regresi_linear_ganda(dataX, dataY)

```python
# Gluck

dataX = {'x1' : [12, 14, 10, 16, 18, 24, 12], 'x2' : [10, 11, 14, 13, 15, 20, 8]}
dataY = [6, 7, 8, 8, 9, 10, 5]

# nomor 1
for key, values in dataX.items():
    if key == 'x1':
        rata_x1 = sum(values) / len(values)
print("1. Hitung Rata - Rata Kompetensi Karyawan (ğ‘‹1) :", rata_x1)

# nomor 2
kuadrat_x1 = []
for key, values in dataX.items():
    if key == 'x1':
        for i in values:
            kuadrat_x1.append(i ** 2)
kuadrat_x1 = sum(kuadrat_x1)
print("2. Hitung Total Kuadrat Kompetensi Karyawan (ğ‘‹1) :", kuadrat_x1)

# nomor 3
for key, values in dataX.items():
    if key == 'x1':
        hasil_x1 = kuadrat_x1 - ((sum(values) ** 2) / len(values))
print("3. Hitung Rata - Rata Kompetensi Karyawan (ğ‘‹1) :", hasil_x1)

# nomor 4
for key, values in dataX.items():
    if key == 'x2':
        rata_x2 = sum(values) / len(values)
print("4. Hitung Rata - Rata Kompetensi Karyawan (ğ‘‹2) :", rata_x2)

# nomor 5
kuadrat_x2 = []
for key, values in dataX.items():
    if key == 'x2':
        for i in values:
            kuadrat_x2.append(i ** 2)
kuadrat_x2 = sum(kuadrat_x2)
print("5. Hitung Total Kuadrat Kompetensi Karyawan (ğ‘‹2) :", kuadrat_x2)

# nomor 6
for key, values in dataX.items():
    if key == 'x2':
        hasil_x2 = kuadrat_x2 - ((sum(values) ** 2) / len(values))
print("6. Hitung Rata - Rata Kompetensi Karyawan (ğ‘‹2) :", hasil_x2)

# nomor 7
rata_y = sum(dataY) / len(dataY)
print("7. Hitung Rata - Rata Kinerja Karyawan (ğ‘Œ) :", rata_y)

# nomor 8
kali_x1_y = []
for key, values in dataX.items():
    if key == 'x1':
        for a, b in zip(values, dataY):
            kali_x1_y.append(a * b)
kali_x1_y = sum(kali_x1_y)
print("8. Hitung hasil kali (ğ‘‹1 ğ‘Œ) :", kali_x1_y)

# nomor 10
sum_dataY = sum(dataY)
hasil_x1_y = []
for (key, values), y in zip(dataX.items(), dataY):
    if key == 'x1':
        hasil_x1_y = kali_x1_y - ((sum(values)*sum_dataY) / len(values))
# kali_x1_y = sum(kali_x1_y)
print("10. Hitung (x1 ğ‘Œ) :", hasil_x1_y)

# nomor 11
kali_x2_y = []
for key, values in dataX.items():
    if key == 'x2':
        for a, b in zip(values, dataY):
            kali_x2_y.append(a * b)
kali_x2_y = sum(kali_x2_y)
print("11. Hitung hasil kali (ğ‘‹2 ğ‘Œ) :", kali_x2_y)

# nomor 12
sum_dataY = sum(dataY)
hasil_x2_y = []
for (key, values), y in zip(dataX.items(), dataY):
    if key == 'x2':
        hasil_x2_y = kali_x2_y - ((sum(values)*sum_dataY) / len(values))
# kali_x1_y = sum(kali_x1_y)
print("12. Hitung (x1 ğ‘Œ) :", hasil_x2_y)

#nomor 13
kali_x1_x2 = []
# for key, values in dataX.items():
for a, b in zip(dataX['x1'], dataX['x2']):
    kali_x1_x2.append(a * b)
kali_x1_x2 = sum(kali_x1_x2)
print("13. Hitung Hasil kali (x1 x2) :", kali_x1_x2)

# nomor 14
for key, values in dataX.items():
    if key == 'x1':
        sum_x1 = sum(values)
    if key == 'x2':
        sum_x2 = sum(values)
hasil_x1_x2 = kali_x1_x2 - ((sum_x1*sum_x2) / len(dataY))
print("14. Hitung (x1 x2) :", hasil_x1_x2)

# nomor 15
koefisien_b1 = ((hasil_x2)*(hasil_x1_y)-(hasil_x1_x2)*(hasil_x2_y)) / ((hasil_x1)*(hasil_x2)-(hasil_x1_x2**2))
koefisien_b1 = round(koefisien_b1, 5)
print("15. Hitung koefisien ğ‘1 :", koefisien_b1)

# nomor 16
koefisien_b2 = ((hasil_x1)*(hasil_x2_y)-(hasil_x1_x2)*(hasil_x1_y)) / ((hasil_x1)*(hasil_x2)-(hasil_x1_x2**2))
koefisien_b2 = round(koefisien_b2, 6)
print("16. Hitung koefisien ğ‘2 :", koefisien_b2)

# nomor 17
koefisien_b0 = rata_y - (koefisien_b1*rata_x1) - (koefisien_b2*rata_x2)
koefisien_b0 = round(koefisien_b0, 6)
print("17. Hitung koefisien ğ‘0 :", koefisien_b0)

# nomor 18
# Y = 2.098049 - 0.01564X1 + 0.439034X2
print(f"18. Buat persamaan regresi gandanya : Y = {koefisien_b0} {koefisien_b1}Xâ‚ + {koefisien_b2}Xâ‚‚")
```
